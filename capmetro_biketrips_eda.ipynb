## Trip Duration

```python
#Get statistics table of trip duration 
df[['trip_duration_minutes']].describe()
```

The table below shows the summary statistics for the trip durations of Capital Metro Bikes:

- **Mean**: The average trip duration is 27.06 minutes. This indicates that, on average, trips are around 27 minutes long.
- **Standard Deviation**: The standard deviation is 70.25 minutes, showing a high level of variability in trip durations. This indicates that there are many trips that vary significantly in length, with a few being much longer or shorter than the mean.
- **Minimum**: The shortest trip duration is 2 minutes, indicating that there are very short trips recorded.
- **25th Percentile (Q1)**: The first quartile (Q1) is 6 minutes, meaning 25% of the trips are 6 minutes or shorter.
- **50th Percentile (Median)**: The median trip duration is 11 minutes, meaning half of the trips are shorter than 11 minutes and half are longer.
- **75th Percentile (Q3)**: The third quartile (Q3) is 27 minutes, meaning 75% of trips are 27 minutes or shorter.
- **Maximum**: The longest trip recorded lasts 1,439 minutes (about 1 day), which is likely an outlier compared to the rest of the trip durations.

**Interpretation:**

The data shows a large amount of variability in bike trip durations. The distribution is highly skewed, with a few very long trips inflating the mean. The range is 1,437 minutes (about 1 day), and the standard deviation is 70 minutes, which is quite high compared to the mean of 27 minutes. This indicates that the trip durations vary widely, and while the average trip duration is around 27 minutes, the median is much lower at 11 minutes, suggesting that most trips are relatively short. The minimum trip duration is 2 minutes, pointing to a substantial number of very brief trips as well. The large standard deviation further supports the idea of significant variability, likely due to a small number of long outlier trips.

```python
#Display
sns.histplot(df['trip_duration_minutes'])
plt.title('Distribution of Capital Metro Bike Trip Durations (in Minutes)')
plt.show()
```

```python
#Display
sns.boxplot(df['trip_duration_minutes'])
plt.title('Capital Metro Bike Trip Duration Distribution (Boxplot)')
plt.show()
```

**As seen in the statistics table and the two visualizations, there is significant variability in the trip durations for Capital Metro Bikes. To address this and better visualize the distribution, I will apply a log transformation and replot the histogram and boxplot.**

```python
#Apply log transformation on trip_duration 
import numpy as np
df['log_trip_duration'] = np.log(df['trip_duration_minutes'])
```

```python
#Get stats table
df[['log_trip_duration']].describe()
```

- **Mean**: The average log-transformed trip duration is 2.57, which corresponds to an original trip duration of about 13 minutes (since the log of 13 is approximately 2.57).
- **Standard Deviation**: The standard deviation is 1.07, indicating moderate variability in the log-transformed data.
- **Minimum**: The shortest trip duration, after transformation, is 0.69, which corresponds to an original trip duration of about 2 minutes (since the log of 2 is approximately 0.69).
- **25th Percentile (Q1)**: 1.79, corresponding to an original trip duration of about 6 minutes.
- **50th Percentile (Median)**: 2.40, corresponding to an original trip duration of about 11 minutes.
- **75th Percentile (Q3)**: 3.30, corresponding to an original trip duration of about 27 minutes.
- **Maximum**: The longest trip duration in the log-transformed data is 7.27, which corresponds to an original trip duration of about 1,440 minutes (or 1 day).

**Interpretation:**

The log transformation has reduced the skewness of the data and compressed the range of trip durations. The mean log-transformed trip duration (2.57) is much closer to the median (2.40), indicating that the data is now more normally distributed now. The transformation also reduces the impact of extreme values, such as the very long trip durations, which were disproportionately influencing the original distribution.

```python
#Redo histogram
sns.histplot(df['log_trip_duration'], bins=10)
plt.title('Distribution of Capital Metro Bike Trip Durations (in Minutes)')
plt.show()
```

```python
#Redo boxplot
#Display
sns.boxplot(df['log_trip_duration'])
plt.title('Capital Metro Bike Trip Duration Distribution (Boxplot)')
plt.show()
```

## Month

```python
#Apply function to return the names of each month
def month_names(x):
    if x == 1:
        return 'January'
    elif x == 2:
        return 'Feburary'
    elif x == 3:
        return 'March'
    elif x == 4:
        return 'April'
    elif x == 5:
        return 'May'
    elif x == 6:
        return 'June'
    elif x == 7:
        return 'July'
    elif x == 8:
        return 'August'
    elif x == 9:
        return 'September'
    elif x == 10:
        return 'October'
    elif x == 11:
        return 'November'
    else:
        return 'December'
#Apply function to the months column
df['month'] = df['month'].apply(month_names)

#Check result
num_unique_months = df['month'].nunique() #Should still be 12 unique month values
print(f'Number of unique months: {num_unique_months}')
print(list(df['month'].unique()))
```

```python
#Get value counts
months = df['month'].value_counts()
months
```

```python
#Display
plot = sns.barplot(months, palette='mako', orient='h')
plt.title('Frequency of Cap Metro Bike Trips by Month')
#Add numbers to bars
for i in plot.containers:
    plot.bar_label(i, )
#Tilt y-axis
plt.yticks(rotation=35)
plt.ylabel('Months')
plt.xlabel('Frequency')
plt.show()
```

```python
#Apply function to classify seasons
def seasons(x):
    if x in ['March', 'April', 'May']:
        return 'Spring'
    elif x in ['June', 'July', 'August']:
        return 'Summer'
    elif x in ['September','October','November']:
        return 'Fall'
    else:
        return 'Winter'
#Apply function
df['season'] = df['month'].apply(seasons)

#Check value counts
season = df['season'].value_counts()
season
```

```python
#Display
plot = sns.barplot(season, palette='mako')
plt.title('Seasonal Distribution of Cap Metro Bike Trips')
#Add numbers to bars
for i in plot.containers:
    plot.bar_label(i, )
plt.xlabel('Seasons')
plt.ylabel('Frequency')
plt.show()
```

# Cap Metro Bike Usage: A Data-Driven Story

## Membership Trends

Cap Metro Bike usage is driven by students and frequent local riders. The **Student Membership** leads all categories with over **430,000 trips**, followed closely by **Local365** (389,000) and **UT Student Membership** (289,000), reflecting strong demand from the University of Texas community and nearby residents.

Short-term passes like the **24 Hour Walk Up Pass** (272,000) and **Walk Up** (260,000) are also popular, indicating a sizable group of occasional riders. In contrast, other options like **Explorer**, **Local31**, and **Weekender** see significantly lower usage, with under 125,000 trips each.

**Key Insight**: Long-term and student-centered memberships dominate, highlighting the program’s core rider base — students and regular commuters.

---

## Classic vs. Electric: What Are Riders Choosing?

Classic bikes account for nearly **1.35 million trips**, significantly ahead of electric bikes at **923,000**. This gap suggests that, while electric options are growing in popularity, classic bikes remain the preferred mode — likely due to cost, accessibility, or habit.

**Key Insight**: Despite modernization, traditional bikes remain central to the system’s usage.

---

## UT at the Center: Checkout Patterns

Checkout activity is heavily centered around UT Austin. The busiest kiosk, **21st & Speedway @ PCL**, logged nearly **180,000 checkouts**, far outpacing others. Additional UT-adjacent stations like **Dean Keeton & Speedway**, **28th & Rio Grande**, and **21st & Guadalupe** also report strong usage, each exceeding 60,000 checkouts.

This concentration confirms that the UT campus and its surroundings form the heart of the Cap Metro network.

---

## Return Patterns Reflect Round-Trip Use

Return data closely mirrors checkout activity. The same UT-area kiosks dominate returns — **21st & Speedway @ PCL** again leads with over **183,000 returns**, followed by **Dean Keeton & Speedway** (103,000), and **28th & Rio Grande** (61,000). The alignment of checkouts and returns suggests that many trips are short, round-trip rides.

**Key Insight**: Most riders are likely commuting within the UT area or making quick returns, underlining the importance of bike availability in high-demand zones.

---

## Trip Duration

Although the **average trip lasts 27 minutes**, the **median is only 11 minutes**, indicating that most trips are quick. A small number of longer rides skew the average, but the majority fall within a 6–27 minute window.

**Key Insight**: The system serves short-distance travel needs — optimizing for quick turnover and bike availability is crucial.

---

## Seasonality & Weather Patterns

Bike usage peaks in **Spring (775,000 trips)** and **Fall (639,000)** — especially during **March** and **April**, each exceeding 260,000 trips. **October** and **September** also see elevated ridership.

In contrast, **Summer (461,000)** and **Winter (393,000)** see declines. Hot Texas summers and holiday-season breaks likely contribute to this trend.

**Key Insight**: Ridership follows academic and weather cycles. Planning for seasonal demand — and off-season maintenance — is vital.

---

## Weekly Patterns: More Than Just a Weekday Commute

Weekends drive significant ridership. **Saturday** leads with nearly **400,000 trips**, followed by **Sunday (346,000)** and **Friday (344,000)**. This suggests strong demand for leisure use and weekend transportation.

Weekdays still show robust use — especially **Monday through Thursday**, which all log between 287,000 and 304,000 trips.

**Key Insight**: Cap Metro Bikes are more than a commuting tool — they’re also a go-to option for weekend recreation.

---

## Hourly Trends: The Afternoon Surge

The highest volume of trips occurs during the late afternoon. The peak hour is **5 PM (190,000+ trips)**, followed closely by **4 PM and 3 PM**, each with over 180,000. Usage steadily builds from **11 AM** onward and drops after **9 PM**, with minimal activity between **3–6 AM**.

**Key Insight**: The system must be optimized for peak demand from noon through early evening — especially around 3 to 6 PM.

---

## Time Periods: When Riders Roll

- **Lunch Break (11 AM–2 PM)** is the most active period, with nearly **498,000 trips**.
- **Evening Leisure (6–9 PM)** follows with **428,000** trips.
- **Evening Rush (4–6 PM)** adds another **375,000** trips.
- **Afternoon Surge (2–4 PM)** rounds out the high-use blocks with **359,000** trips.

The **Morning Commute (7–9 AM)** sees the fewest trips — just **113,000** — suggesting that bikes are underutilized during early hours.

**Key Insight**: Midday and post-work periods dominate usage. Morning commutes are a growth opportunity.

---

## Final Insights

Across all dimensions — time, location, and rider type — Cap Metro Bikes support **short, flexible trips** centered around UT Austin. Peak usage aligns with **student schedules**, **mild weather**, and **afternoon hours**, while weekends and leisure time drive strong ridership.

Understanding these patterns enables smarter fleet management, station placement, and promotional strategies to improve service and expand reach.
